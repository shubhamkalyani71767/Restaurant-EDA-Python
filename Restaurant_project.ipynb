```python
#Load the Data and import the libarary
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv(r'C:\Users\shubham kalyani\OneDrive\Desktop\study material\Machine learning\dataset\analysis\restaurant.csv')
df
```

```python
df.describe(include=['float','int'])
```

```python
df.describe(include=['object'])
```

```python
# Total Revenue
total_revenue = df['total_bill'].sum()

# Total Tips
total_tips = df['tip'].sum()

# Average Order Value
avg_order = df['total_bill'].mean()

# Average Tip %
df['tip_percent'] = (df['tip'] / df['total_bill']) * 100
avg_tip_percent = df['tip_percent'].mean()

print("Total Revenue:", total_revenue)
print("Total Tips:", total_tips)
print("Average Order Value:", avg_order)
print("Average Tip %:", avg_tip_percent)
```

```python
revenue_by_day = df.groupby('day')['total_bill'].sum().sort_values(ascending=False)
print(revenue_by_day)
```

```python
revenue_by_time = df.groupby('time')['total_bill'].sum()
print(revenue_by_time)
```

```python
orders_by_day = df['day'].value_counts()
print(orders_by_day)
```

```python
tip_by_gender = df.groupby('gender')['tip_percent'].mean()
print(tip_by_gender)
```

```python
tip_by_smoker = df.groupby('smoker')['tip_percent'].mean()
print(tip_by_smoker)
```

```python
tip_by_day = df.groupby('day')['tip_percent'].mean()
print(tip_by_day)
```

```python
high_bill = df[df['total_bill'] > df['total_bill'].mean()]
print("High Value Orders Count:", high_bill.shape[0])
```

```python
revenue_by_day = df.groupby('day')['total_bill'].sum()

plt.figure()
revenue_by_day.plot(kind='bar')
plt.title("Total Revenue by Day")
plt.xlabel("Day")
plt.ylabel("Total Revenue")
plt.show()
```

```python
revenue_by_time = df.groupby('time')['total_bill'].sum()

plt.figure()
revenue_by_time.plot(kind='bar')
plt.title("Revenue by Meal Time")
plt.xlabel("Meal Time")
plt.ylabel("Total Revenue")
plt.show()
```

```python
avg_bill_by_size = df.groupby('size')['total_bill'].mean()

plt.figure()
avg_bill_by_size.plot(kind='line')
plt.title("Average Bill by Party Size")
plt.xlabel("Party Size")
plt.ylabel("Average Bill")
plt.show()
```

```python
plt.figure()
plt.scatter(df['total_bill'], df['tip'])
plt.title("Total Bill vs Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
orders_by_day = df['day'].value_counts()

plt.figure()
orders_by_day.plot(kind='bar')
plt.title("Number of Orders by Day")
plt.xlabel("Day")
plt.ylabel("Number of Orders")
plt.show()
```
